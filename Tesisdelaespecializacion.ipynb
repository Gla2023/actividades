```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import Voronoi, voronoi_plot_2d

# Define las ubicaciones de los puntos en el plano
points = np.array([[0, 0], [1, 2], [3, 1], [2, 2]])

# Calcula el diagrama de Voronoi
vor = Voronoi(points)

# Calcula las áreas de las regiones de Voronoi
areas = []
for region_index in vor.point_region:
    region = vor.regions[region_index]
    if -1 not in region and region:  # Verifica que no sea una región infinita y esté dentro de los límites
        vertices = vor.vertices[region]
        areas.append(abs(np.cross(vertices[:-1], vertices[1:]).sum()) / 2)  # Calcula el área del polígono
    else:
        areas.append(np.nan)  # Usamos NaN para las regiones infinitas o fuera de los límites

# Visualiza el diagrama de Voronoi
voronoi_plot_2d(vor)
plt.plot(points[:, 0], points[:, 1], 'ko')  # Plotea los puntos

# Muestra las áreas de las regiones
for i, region_index in enumerate(vor.point_region):
    area = areas[i]  # Usamos el índice i en lugar de region_index
    if not np.isnan(area):  # Verifica si el área es un número
        plt.text(vor.points[i, 0], vor.points[i, 1], f"Area: {area:.2f}", fontsize=8)

plt.xlim(-10, 10)  # Límites en el eje x
plt.ylim(-10, 10)  # Límites en el eje y
plt.show()

```
